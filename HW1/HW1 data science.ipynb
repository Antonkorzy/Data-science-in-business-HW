Min
If "March, 13" is the first day with >= 50 confirmed cases, then "March, 22" is the day you need

```python
import pandas as pd

import numpy as np
```

```python
conf_cases = pd.read_csv(
                    'time_series_covid19_confirmed_global.csv',
                    engine="python")
```

```python
del conf_cases["Lat"]
del conf_cases["Long"]
pd.options.display.max_columns = 999
conf_cases_country = conf_cases.groupby("Country/Region").sum()
conf_cases_country
```

```python
def cases_on_day_my(data, min_cases=50, days=9):
    l = []
    for index,row in data.iterrows():
        first_day = -1
        for ind in range(len(row)):
            if row[ind] >= min_cases:
                first_day = ind
                break
        if (first_day + days < len(row)) and first_day != -1:
            l.append(row[first_day+days])
    return l
```

```python
from statistics import median

def print_stat(li):
    max_value = str(max(li))
    min_value = str(min(li))
    avg_value = 0 if len(li) == 0 else sum(li)/len(li)
    median_value = str(median(li))
    print("max_value:" + max_value + '\n' + "min_value: " + min_value + '\n' + 
          "avg_value: " + str(avg_value) + '\n' + "median_value: " + median_value)
```

```python
cases_test = conf_cases_country.T.set_index(pd.to_datetime(conf_cases_country.columns))
cases_test.head(5)
```

```python
print_stat(cases_on_day_my(conf_cases_country))
```

# Task 5.
On which weekday most cases were reported in France on average?
On which weekday least cases were reported in Italy on average?
Write all numbers rounded to 2 decimal places.

```python
france = conf_cases[conf_cases["Country/Region"] == "France"].groupby("Country/Region").sum()
```

```python
france
```

```python
li_count = [0,0,0,0,0,0,0]
li_sum = [0,0,0,0,0,0,0]
for index,row in france.iterrows():
    for ind in range(len(row)-1):
        li_sum[(((ind+1) % 7) + 2) % 7] += (row[ind+1] - row[ind])
        li_count[(((ind+1) % 7) + 2) % 7] += 1
```

```python
for ind in range(0,7):
    print(li_sum[ind] / li_count[ind])
```

```python
italy = conf_cases[conf_cases["Country/Region"] == "Italy"].groupby("Country/Region").sum()
```

```python
li_count = [0,0,0,0,0,0,0]
li_sum = [0,0,0,0,0,0,0]
for index,row in italy.iterrows():
    for ind in range(len(row)-1):
        li_sum[(((ind+1) % 7) + 2) % 7] += (row[ind+1] - row[ind])
        li_count[(((ind+1) % 7) + 2) % 7] += 1
```

```python
for ind in range(0,7):
    print(li_sum[ind] / li_count[ind])
```
