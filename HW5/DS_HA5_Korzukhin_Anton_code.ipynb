```python
import pandas as pd
```

```python
df = pd.read_csv("sample_transations.csv")
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
df = df.join(pd.get_dummies(df['hour']))
df = df.join(pd.get_dummies(df['dbi_item_catgry'], prefix='catgry'))
df = df.join(pd.get_dummies(df['dbi_item_sub_catgry'], prefix='sub_catgry'))
df = df.join(pd.get_dummies(df['dbi_item_famly_name'], prefix='famly_name'))
df = df.join(pd.get_dummies(df['dbi_item_size'], prefix='size'))
```

```python
median_price_size = df.groupby('dbi_item_size')['post_discount_price'].median().reset_index()
df = df.merge(median_price_size, on='dbi_item_size', suffixes=("", 'median_price_per_size'))
df
```

```python
median_price_person = df.groupby('dd_card_number')['post_discount_price'].median()
df = df.merge(median_price_person, on='dd_card_number', suffixes=("", 'median_price_per_person'))
```

```python
median_price_store = df.groupby('store_number')['post_discount_price'].median()
df = df.merge(median_price_store, on='store_number', suffixes=("", 'median_price_per_store'))
```

```python
median_price_store_on_sum = df.groupby('store_number')['post_discount_price'].median() /  df.groupby('store_number')['post_discount_price'].sum()
df = df.merge(median_price_store_on_sum, on='store_number', suffixes=("", 'median_price_store_on_sum'))
```

```python
median_price_person_on_sum = df.groupby('dd_card_number')['post_discount_price'].median() /  df.groupby('dd_card_number')['post_discount_price'].sum()
df = df.merge(median_price_person_on_sum, on='dd_card_number', suffixes=("", 'median_price_person_on_sum'))
```

```python
median_price_person_per_day = df.groupby(['dd_card_number','date'])['post_discount_price'].median().reset_index()
df = df.merge(median_price_person_per_day, on=['dd_card_number','date'], suffixes=("", 'median_price_person_per_day'))
```

```python
df
```

```python
del df['hour']
del df['dbi_item_catgry']
del df['dbi_item_sub_catgry']
del df['dbi_item_famly_name']
del df['dbi_item_size']
```

```python
df['Date'] = pd.to_datetime(df['date'], format='%m/%d/%Y') 
df['year'] = df['Date'].dt.year
df['week'] = df['Date'].dt.week
df['day'] = df['Date'].dt.day
del df['date']
del df['Date']
```

```python
len(df)
```

```python
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

ns = list(range(2, 20, 1))
plt.plot(ns, [KMeans(n_clusters=n).fit(df).score(df) for n in ns]);
```
